```python
def Linear_regression(X , y, alpha, iter):
    normalising_param = 2*X.shape[0]
    theta = np.ones(X.shape[1])                                       
    theta_req = np.ones(X.shape[1])                                 
    h_Theta = np.matmul(X, theta)
    loss = h_Theta - y
    j_Theta_min = np.matmul(loss.transpose(), loss)/normalising_param
    
    for i in range(iter):
        for feature in range(X.shape[1]):
            gradient = np.sum ( np.multiply(loss, X.T[feature]) )/X.shape[0]
            theta[feature] = theta[feature] - alpha*gradient
        h_Theta = np.matmul(X, theta)
        loss = h_Theta - y
        j_Theta = np.matmul(loss.transpose(), loss)/normalising_param
        print(j_Theta)
        if(j_Theta < j_Theta_min):
            j_Theta_min = j_Theta
            theta_req = theta
    
    return theta_req 
    
        
    
```

```python
theta = Linear_regression(X, y, 0.01, 1000)
```
